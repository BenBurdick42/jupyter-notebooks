```python
import numpy as np
from numpy import linalg as la
```

```python
zero = np.array([1,0])

one = np.array([0,1])

Z = np.outer(zero,zero) - np.outer(one,one)

I = np.eye(2)

# print(np.kron(I,Z))

def Z_i(i):
    if i == 0:
        a = np.kron(Z,I)
        b = np.kron(a,I)
        c = np.kron(b,I)
        d = np.kron(c,I)
        return d
    if i == 1:
        a = np.kron(I,Z)
        b = np.kron(a,I)
        c = np.kron(b,I)
        d = np.kron(c,I)
        return d
    if i == 2:
        a = np.kron(I,I)
        b = np.kron(a,Z)
        c = np.kron(b,I)
        d = np.kron(c,I)
        return d
    if i == 3:
        a = np.kron(I,I)
        b = np.kron(a,I)
        c = np.kron(b,Z)
        d = np.kron(c,I)
        return d
    if i == 4:
        a = np.kron(I,I)
        b = np.kron(a,I)
        c = np.kron(b,I)
        d = np.kron(c,Z)
        return d
print(Z_i(2))
```
